# Random Forest Classifier

The lending data from 2007-2010 is used to classify and predict whether or not the borrower paid back their loan in full

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
loans = pd.read_csv('loan_data.csv')
```

```python
loans.info()
```

```python
loans.head(10)
```

### Exploratory Data Analysis


```python
# Creating a histogram of two FICO distributions on top of each other

plt.figure(figsize=(14,6))
loans[loans['credit.policy']==1]['fico'].hist(alpha=0.5,
                                              bins=35,label='Credit.Policy=1')
loans[loans['credit.policy']==0]['fico'].hist(alpha=0.5,
                                              bins=35,label='Credit.Policy=0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(14,6))
loans[loans['not.fully.paid']==1]['fico'].hist(alpha=0.5,color='blue',
                                              bins=30,label='not.fully.paid=1')
loans[loans['not.fully.paid']==0]['fico'].hist(alpha=0.5,color='red',
                                              bins=30,label='not.fully.paid=0')
plt.legend()
plt.xlabel('FICO')
```

```python
#a countplot using seaborn showing the counts of loans by purpose, 
#  with the color hue defined by not.fully.paid
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=loans,palette='Set1')
```

```python
#the trend between FICO score and interest rate
sns.jointplot(x='fico',y='int.rate',data=loans,color='purple')
```

```python
# the trend differed between not.fully.paid and credit.policy
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loans,
           hue='credit.policy',
           col='not.fully.paid',palette='Set1')
```

### Data Feature Engineering

```python
# The purpose column is categorical, hence converting into Dummy variable
final_data = pd.get_dummies(loans,columns=['purpose'],drop_first=True)
```

```python
final_data.info()
```

### Data Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

### Evaluation of Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

### Training the Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=800)
```

```python
rfc.fit(X_train,y_train)
```

```python
#Prediction and Evaluation
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

### Random Forest model has a better score
