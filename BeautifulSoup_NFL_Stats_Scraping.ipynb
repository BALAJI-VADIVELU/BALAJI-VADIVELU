```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.nfl.com/standings/league/2019/reg/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
```

```python
table = soup.find('table', {'data-require':'modules/tableSortable'})
```

```python
table
```

Extracting the column headers

```python
headers = []
for head in table.find_all('th'):
    title = head.text.strip() # Removing the white space
    headers.append(title)
```

```python
headers
```

Extracting the row data

```python
df = pd.DataFrame(columns = headers)
```

```python
for row in table.find_all('tr')[1:]: 
    
    first_td = row.find_all('td')[0].find('div', class_ = 'd3-o-club-fullname').text.strip()
    data = row.find_all('td')[1:]
    row_data = [td.text.strip() for td in data]
    row_data.insert(0,first_td)
    length = len(df)
    df.loc[length] = row_data
```
